# Introduction


This Dataset provides insights into how daily screen time, social media usage, and lifestyle habits affect mental well-being. By analyzing factors such as stress, sleep quality, exercise frequency, and happiness, this dataset helps explore the balance between digital life and overall mental health. It serves as a valuable resource for understanding the impact of technology on emotional and physical well-being in the modern era.

# Create countplot

```python
sns.countplot(x="Stress_Level(1-10)", data=df)
plt.show()
```

#Create violinplot


```python
sns.violinplot(x="Exercise_Frequency(week)",y="Happiness_Index(1-10)",data=df)
plt.show()
```

# Create pie chart

```python
labels = ['User_ID', 'Age', 'Daily_Screen_Time(hrs)', 'Sleep_Quality(1-10)']
sizes = [25, 35, 20, 20]
colors = ['blue', 'yellow', 'pink', 'brown']
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Social Media and Mental Health Balance')
plt.axis('equal')
plt.show()
```

# Create Scatter plot

```python
sns.scatterplot(x = "Days_Without_Social_Media",y = "Social_Media_Platform",data = df,hue = "Gender")
plt.show()
```

# Create lineplot

```python
sns.lineplot(x="Social_Media_Platform",y="Days_Without_Social_Media",data=df)
plt.title("Social Media and Mental Health Balance")
plt.xlabel("Social_Media_Platform")
plt.ylabel("Days_Without_Social_Media")
plt.xticks(rotation = 25)

plt.show()
```

# Create Pair Plot

```python
sns.pairplot(df,hue="Age")
plt.show()
```

# Create HeatMap

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
